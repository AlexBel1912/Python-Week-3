```python
#Merge
query_dimproduct = """SELECT *
        FROM dimproduct
        """
dimproduct = pd.read_sql(query_dimproduct, db_engine)
query_factresellersales = """ SELECT *
            FROM factresellersales
            """
factresellersales = pd.read_sql(query_factresellersales, db_engine)
factresellersales.merge(dimproduct, how="left", on="ProductKey")
```

```python
merged = factresellersales.merge(dimproduct, how="left", on="ProductKey")[['SalesOrderNumber', 'SalesOrderLineNumber', 'OrderDate', 'ProductKey', 'EnglishProductName']]
merged
```

```python
import pandas as pd 
data = { 
    "Giorno": ["Lunedì", "Martedì", "Mercoledì", "Giovedì", "Venerdì"], 
    "Vendite": [100, 200, 150, 300, 250] 
} 
df = pd.DataFrame(data) 
df.loc[:, "Vendite Cumulative"] = df.loc[:, "Vendite"].cumsum() 
print(df)
```
