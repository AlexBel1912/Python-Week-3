Esercizio 
Il dataset insurance.csv dei beginner_datasets contiene dati su caratteristiche e abitudini delle persone rispetto ai costi individuali 
per le cure mediche come premio per le assicurazioni sulla salute. 
• Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
• Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 
• E rispetto a smoker? E a sex? 
• Quali sono i descrittori statistici di bmi? 
Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi? 
Nota: è lo stesso esercizio della volta scorsa, ma stavolta da eseguire con il metodo .groupby()

```python
insurance = pd.read_csv("insurance.csv")
numero_righe, numero_colonne = insurance.shape
print(f"Il dataset contiene {numero_righe} righe e {numero_colonne} colonne." )
display(insurance)

print("\nNomi delle colonne:")
print(insurance.columns)


print(f"Dimensioni del dataset: {insurance.shape}")
print("\nPrime 5 righe del dataset:")
print(insurance.head())
```

Quali sono le medie di charges rispetto a region? Ci sono differenze significative?

```python
media_charges_region = insurance.groupby('region')['charges'].mean().round(2)
print("\nMedie di charges rispetto a region:")
print(media_charges_region)

deviazioni_charges_region = insurance.groupby('region')['charges'].std()
print("\nDeviazioni standard di charges rispetto a region:")
print(deviazioni_charges_region)
```

• E rispetto a smoker? E a sex?

```python
media_charges_smoker = insurance.groupby('smoker')['charges'].mean().round(2)
print("\nMedie di charges rispetto a smoker:")
print(media_charges_smoker)

media_charges_sex = insurance.groupby('sex')['charges'].mean().round(2)
print("\nMedie di charges rispetto a sex:")
print(media_charges_sex)
```

Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?

```python
print("\nDescrittori statistici di bmi:")
print(insurance['bmi'].describe())

quartili_bmi = insurance['bmi'].quantile([0.25, 0.5, 0.75])
print("\nQuartili di bmi:")
print(quartili_bmi)

# Analisi di charges rispetto ai quartili di bmi
def analisi_charges_per_quartile(bmi):
    if bmi <= quartili_bmi[0.25]:
        return 'Q1'
    elif bmi <= quartili_bmi[0.5]:
        return 'Q2'
    elif bmi <= quartili_bmi[0.75]:
        return 'Q3'
    else:
        return 'Q4'

insurance['bmi_quartile'] = insurance['bmi'].apply(analisi_charges_per_quartile)

# Calcola le statistiche di charges per quartile di bmi 
statistiche_charges_bmi = insurance.groupby('bmi_quartile')['charges'].agg(['min', 'mean', 'max']).round(2)
print("\nStatistiche di charges rispetto ai quartili di bmi:")
print(statistiche_charges_bmi)
```

Esercizio 
Carichiamo in un DataFrame il dataset iris.csv dei beginner_datasets e calcoliamo: 
• La media della lunghezza dei petali di tutto il dataset 
• La media della lunghezza dei petali per ogni specie di Iris, utilizzando il metodo .groupby() 
• Media, minimo e massimo della larghezza dei sepali per ogni specie, utilizzando .groupby() e .agg()

```python
iris = pd.read_csv('iris.csv')
display(iris)
```

```python
medie_petal_length = iris['petal_length'].mean().round(2)
print(f"\nMedia lunghezza petali: {medie_petal_length}")
```

```python
medie_petal_length_species = iris.groupby('species')['petal_length'].mean().round(2)
print(f"\nLa media della lunghezza dei petali per ogni specie di Iris:")
print(medie_petal_length_species)
```

```python
statistiche_sepal_width_species = iris.groupby('species')['sepal_width'].agg(['mean', 'min', 'max'])
print("\nStatistiche larghezza sepali per specie:")
print(statistiche_sepal_width_species)
```

```python
# Read the wine dataset from CSV file into a pandas DataFrame
wine = pd.read_csv('wine.csv')
# Display the first 5 rows of the DataFrame to preview the data
wine.head()
```

```python
# Calculate mean alcohol concentration grouped by wine type
media_alcol_type = wine.groupby('type')['alcohol'].mean()

# Print header for the results
print("\nMedia concentrazione alcolica per vini bianchi e vini rossi")

# Display mean alcohol concentration for white and red wines
print(media_alcol_type)
```

```python
# Calculate the mean pH value for each wine type (white and red)
media_pH_type = wine.groupby('type')['pH'].mean()

# Print header for the pH means
print("\nMedia ph per vini bianchi e vini rossi")
# Display the mean pH values grouped by wine type
print(media_pH_type)
```

```python
media_sulphates_type = wine.groupby('type')['sulphates'].mean()
print("\nMedia solfati per vini bianchi e vini rossi")
print(media_sulphates_type)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
boston = pd.read_csv('boston.csv')
print(boston.head())
display(boston)
```

```python
print(boston.describe())
```

```python
# Calculate mean MEDV values grouped by CHAS (Charles River dummy variable)
media_medv_chas = boston.groupby('chas')['medv'].mean()
# Print header for the results
print("\nMedia medv per chas:")
# Display mean MEDV values for each CHAS category (0=no river boundary, 1=bounds river)
print(media_medv_chas)
```

```python
# Calculate the correlation coefficient between crime rate (CRIM) and median house value (MEDV)
correlazione_crim_medv = boston['crim'].corr(boston['medv'])
print(f"\nCorrelazione CRIM-MEDV: {correlazione_crim_medv}")

# Create a scatter plot to visualize the relationship between crime rate and house values
sns.scatterplot(x='crim', y='medv', data=boston)
plt.show()
```

```python
# Calculate mean number of rooms (rm) grouped by pupil-teacher ratio (ptratio)
media_rm_ptratio = boston.groupby('ptratio')['rm'].mean()
# Calculate mean house value (medv) grouped by pupil-teacher ratio (ptratio)
media_medv_ptratio = boston.groupby('ptratio')['medv'].mean()
# Print average rooms per pupil-teacher ratio
print("\nMedia rm per ptratio e")
print(media_rm_ptratio)

# Print average house value per pupil-teacher ratio
print("\nMedia medv per ptratio e")
print(media_medv_ptratio)

sns.scatterplot(x='ptratio', y='rm', data=boston)
plt.show()
sns.scatterplot(x='ptratio', y='medv', data=boston)
plt.show()
```

```python
# Calculate mean values grouped by 'rad' (accessibility to radial highways)
# Calculate average age per rad value
media_age_rad = boston.groupby('rad')['age'].mean()
# Calculate average number of rooms per rad value
media_rm_rad = boston.groupby('rad')['rm'].mean()
# Calculate average tax per rad value 
media_tax_rad = boston.groupby('rad')['tax'].mean()

# Print the results
print("\nMedia age per rad:")
print(media_age_rad)
print("\nMedia rm per rad:")
print(media_rm_rad)
print("\nMedia tax per rad:")
print(media_tax_rad)
```
